### Heatmaps

####  Heatmap by Hour and Day

```python
data_agg_day_hour = data.groupby(['Day','Hour']).size().unstack(level=0)
```

```python
data_agg_day_hour
```

```python
data_agg_day_hour = data_agg_day_hour.reindex(index=sorted(data_agg_day_hour.index, reverse=True))
```

```python
plt.figure(figsize=(14,8))

sns.heatmap(data=data_agg_day_hour,annot=False)

plt.title('Heatmap of rides for different days of month and hours of day')

plt.tight_layout()

plt.show()
```

Trips using Skyline between July-September 2014 peaked at 20-21h and were at their lowest between 0-5am.

#### HeatMap by Month and Day

```python
data_agg_month_day = data.groupby(['Day','Month']).size().unstack(level=0)
```

```python
data_agg_month_day = data_agg_month_day.reindex(index=sorted(data_agg_month_day.index,reverse=True))
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(data_agg_month_day)

plt.title('Heat map by month and day')

plt.show()
```

Scale is very small here, so the colors may not be reliable. Let's examine the distribution of data for each month.

#### Heat Map of Month and Day of Week

```python
data_agg_month_weekday = data.groupby(['DayOfWeek','Month']).size().unstack(level=0)
```

```python
data_agg_month_weekday = data_agg_month_weekday.reindex(index=sorted(data_agg_month_weekday.index,reverse=True))
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data_agg_month_weekday)
plt.title('Heat map by month and day of week')
plt.show()
```

This heatmap confirms the previously mentioned low usage on weekends and higher usage on Tuesday, Wednesday, and Thursday. Additionally, there is more data for July than for August and even more for September, which explains the color gradients.
