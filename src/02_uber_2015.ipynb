### Analysing pickups with different dispatching and affiliated base

```python
diff_base = data[data['Dispatching_base_num']!= data['Affiliated_base_num']]
```

```python
len(diff_base)
```

```python
len(data) - len(diff_base)
```

There is 3647483 pickups where dispaching base num is different from affiliated base num.

```python
diff_base
```

It would be interesting to see weather there is a pattern in those pickups, and try to find the reason for change of base number, because this data represents 1/4 of the whole dataset, it is large amount of data.

We could see that there are only 8 dispaching bases, and a lot more of affiliated bases. Let's see if there is some pattern in mapping between bases.


```python
data_agg_dis_aff_bases = diff_base.groupby(['Dispatching_base_num','Affiliated_base_num']).size().reset_index(name='count')
```

```python
data_agg_dis_aff_bases
```

```python
data_agg_dis_aff_bases[data_agg_dis_aff_bases['Dispatching_base_num']=='B02835']
```

```python
unique_bases = np.unique(data_agg_dis_aff_bases['Dispatching_base_num'])
```

```python
plt.figure(figsize=(12, 6))
for i,base in enumerate(unique_bases):
    curr_data = data_agg_dis_aff_bases[data_agg_dis_aff_bases['Dispatching_base_num']==base]
    top3 = curr_data.groupby('Dispatching_base_num').apply(lambda x: x.nlargest(3, 'count')).reset_index(drop=True)

    plt.subplot(2,4,i+1)

    for base in top3['Dispatching_base_num'].unique():
        subset = top3[top3['Dispatching_base_num'] == base]
        plt.bar(subset['Affiliated_base_num'], subset['count'], label=f"Dispatch Base {base}")

    plt.ylabel('Count')
    plt.title(f'Top 3 Affiliated Bases for {base}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Here we represented top 3 affiliated bases for each of our Dispatching bases.
- We can see that scales differ a lot, ex B02764 is affiliated with B02682 close to 300 000 times while B02835 is affiliated with B02682 around 1300 times. We need to take that in consideration.
- Next thing we can notice is that all of dispaching bases are mostly affiliated with one another, except B02788 which affiliates B02764 around 50 000 times.
- Also we can notice that bases B02682 and B02764 affiliate most of other dispaching bases.

```python
len(data[(data['Month']=='June') & (data['Day']==27)])
```

Looking at this heatmap we cannot make any valuable conclusions, we can't catch any pattern, but we can see that for some days have significantly less trip informations. Such as 2nd, 4th  and 27th of January, while some such as 27th of June and 16th of May have significantly more trip information than others. 
Possible reasons for these numbers can be found if we investigate what happened then:
1. For 2nd January there are no special events that happened, but it's date around hollidays so probably that's the reason for less Uber rides.
2. On January 4, 2015 in New York, various events were noted, including Mike Huckabee's departure from Fox News ahead of a potential run for president in 2016. Also, other news stories took place during that period, such as the death of Edward Brooke, the first elected black senator, as well as the protests and police response to Mayor Bill DeBlasi.
3. On January 27, 2015 in New York, a powerful snowstorm dubbed "Snowmageddon 2015" happened. Travel bans were in place in the city, resulting in empty streets, although some parts of the region saw heavy snowfall, particularly up to 15 inches on Long Island.
4. on June 27th a man was fatally shot and stabbed in Manhattan. Possibly people felt unsecure and took more uber rides.
5. On May 16, 2015, a match between New York City FC and Chicago Fire was held in New York City. Also on that day was the annual gala benefiting the Kristen Ann Carr Fund, featuring Bruce Springsteen with the Tribeca Playboys. Two big events led to more uber rides than usual.
